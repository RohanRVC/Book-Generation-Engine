```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# Read the genre data from the uploaded file
with open('C:/Users/RohanRVC/Documents/Business/E-books/E_Books_Unknown_AI/genre.txt', 'r') as f:
    genres = f.readlines()

# Remove newline characters and other unnecessary whitespace
genres = [x.strip() for x in genres]

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')

# Generate TF-IDF matrix
tfidf = tfidf_vectorizer.fit_transform(genres)

# Initialize NMF
nmf = NMF(n_components=10, random_state=42)

# Apply NMF
nmf_features = nmf.fit_transform(tfidf)

# Get the list of words corresponding to the features
feature_names = tfidf_vectorizer.get_feature_names_out()

# Display the top words for each topic (i.e., main genre)
top_words = {}
for i, component in enumerate(nmf.components_):
    sorted_idx = component.argsort()[:-11:-1]
    top_words[f"Topic {i + 1}"] = [feature_names[i] for i in sorted_idx]

top_words
```

```python
# Define a mapping function to simplify genres based on the provided list of required genres.
required_genres = [
    "Action", "Adventure", "Biography", "Children", "Comedy", "Crime", "Drama", 
    "Fantasy", "History", "Horror", "Mystery", "Poetry", "Romance", "Sci-Fi", "Thriller"
]

def map_genre_to_required(genre_text):
    # Initialize an empty list to store the mapped genres
    mapped_genres = []
    
    # Loop through each required genre and check if it appears in the original genre_text
    for required_genre in required_genres:
        if required_genre.lower() in genre_text.lower():
            mapped_genres.append(required_genre)
    
    # If we found any matches, return them as a comma-separated string
    if mapped_genres:
        return ', '.join(mapped_genres)
    else:
        return "Other"

# Test the mapping function
sample_genres = [
    "Action-packed adventures",
    "Biographical novel",
    "Children's literature",
    "Comedy and drama",
    "Crime thriller",
    "Historical drama",
    "Horror fiction",
    "Mystery novel",
    "Romantic poetry",
    "Science fiction",
    "Thriller, Action, Adventure"
]

# Apply the mapping function to the sample genres
mapped_sample_genres = [map_genre_to_required(genre) for genre in sample_genres]
mapped_sample_genres
```

```python
#THis will take 3 min 30 sec to run

from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the Word2Vec model
model = KeyedVectors.load("C:/Users/RohanRVC/Documents/Business/E-books/E_Books_Unknown_AI/word2vec-google-news-300.model")




# Create vectors for predefined genres
predefined_genre_vectors = {}
for genre in predefined_genres:
    tokens = genre.lower().split()
    vectors = [model[token] for token in tokens if token in model]
    if vectors:
        avg_vector = np.mean(vectors, axis=0)
        predefined_genre_vectors[genre] = avg_vector

# Function to categorize genre
def categorize_genre(input_genre):
    tokens = input_genre.lower().split()
    vectors = [model[token] for token in tokens if token in model]
    if not vectors:
        return "Other"
    avg_vector = np.mean(vectors, axis=0)
    similarities = {}
    for genre, vector in predefined_genre_vectors.items():
        sim = cosine_similarity([avg_vector], [vector])[0][0]
        similarities[genre] = sim
    return max(similarities, key=similarities.get)

# Categorize genres in your dataset
df_new['Cleaned_Genre'] = df_new['Original_Genre'].apply(categorize_genre)
```

```python
import re

def clean_content(text_content):
    # Remove the Gutenberg metadata at the start and end of the book
    start_keyword = "*** START OF THE PROJECT GUTENBERG EBOOK"
    end_keyword = "*** END OF THE PROJECT GUTENBERG EBOOK"
    
    start_index = text_content.find(start_keyword)
    end_index = text_content.find(end_keyword)
    
    # If both the start and end keywords are found, extract the main content
    if start_index != -1 and end_index != -1:
        main_content = text_content[start_index + len(start_keyword):end_index].strip()
    else:
        main_content = text_content
    
    # Remove any remaining headers or footers
    main_content = re.sub(r"\*\*\*.*?\*\*\*", "", main_content)
    
    # Replace multiple white spaces with a single space
    main_content = re.sub(r"\s+", " ", main_content).strip()
    
    return main_content

# Read the content of the uploaded text file
# with open('path/to/your/content.txt', 'r', encoding='utf-8') as file:
#     text_content = file.read()

# Clean the content
cleaned_content = clean_content(df_new.Content.iloc[4])

# Save the cleaned content to a new file
with open('cleaned_content.txt', 'w', encoding='utf-8') as file:
    file.write(cleaned_content)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Initialize the stopwords and stemmer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def clean_contentss(text):
    # Remove special characters and numbers
    cleaned_text = re.sub(r"[^a-zA-Z0-9\s.,-:]", " ", text)

    
    # Convert to lowercase
    cleaned_text = cleaned_text.lower()
    
    # Tokenization (split the text into words)
    tokens = cleaned_text.split()
    
    # Remove stopwords and perform stemming
    cleaned_tokens = [stemmer.stem(word) for word in tokens if not word in stop_words]
    
    # Joining the cleaned tokens back into a string
    cleaned_text = ' '.join(cleaned_tokens)
    
    return cleaned_text

# Apply the function to your content column
# Assuming df is your DataFrame and 'Content' is your column with the text
# df['Cleaned_Content'] = df['Content'].apply(clean_content)

with open('cleaned_content_nlp.txt','w',encoding='utf-8')as f:
 f.write('')

with open('cleaned_content_nlp.txt','a',encoding='utf-8')as f:
 f.write(clean_contentss(df_new.Content.iloc[4]))
```
