## Now Lets Categorize the Genre

```python
#THis will take 3 min 30 sec to run

from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the Word2Vec model
model = KeyedVectors.load("C:/Users/RohanRVC/Documents/Business/E-books/E_Books_Unknown_AI/word2vec-google-news-300.model")


predefined_genres = ['Fiction',
'Mystery',
'Thriller',
'Romance',
'Science Fiction',
'Fantasy',
'Horror',
'Historical Fiction',
'Contemporary Literature',
'Young Adult',
'Children\'s Books',
'Picture Books',
'Poetry',
'Biography',
'Autobiography',
'Memoir',
'Self-help',
'Business and Finance',
'Travel',
'History',
'Science and Technology',
'Psychology',
'Philosophy',
'Religion and Spirituality',
'Art and Photography',
'Political Science',
'Social Science',
'Essays and Literary Criticism',
'Parenting and Family',
'Nature and Environment',
'True Crime',
'Linguistics and Languages',
'Cultural and Ethnic Studies',
'Drama and Performing Arts',
'Journalism and Media',
'Fashion and Style',
'Philosophy of Mind',
'Mindfulness and Meditation',
'Social Justice and Activism',
'Gardening and Horticulture',
'Food and Cooking',
'Health and Fitness',
'Comics and Graphic Novels',
'Mythology and Folklore',
'Education',
'Science and Nature',
'Reference',
'Sports and Recreation']



# Create vectors for predefined genres
predefined_genre_vectors = {}
for genre in predefined_genres:
    tokens = genre.lower().split()
    vectors = [model[token] for token in tokens if token in model]
    if vectors:
        avg_vector = np.mean(vectors, axis=0)
        predefined_genre_vectors[genre] = avg_vector

# Function to categorize genre
def categorize_genre(input_genre):
    tokens = input_genre.lower().split()
    vectors = [model[token] for token in tokens if token in model]
    if not vectors:
        return "Other"
    avg_vector = np.mean(vectors, axis=0)
    similarities = {}
    for genre, vector in predefined_genre_vectors.items():
        sim = cosine_similarity([avg_vector], [vector])[0][0]
        similarities[genre] = sim
    return max(similarities, key=similarities.get)

# Categorize genres in your dataset
# df_new['Cleaned_Genre'] = df_new['Original_Genre'].apply(categorize_genre)
```

```python
def break_into_chunks(sentence, min_words=10, max_words=25):
    # texts = re.sub(r'\(\d+\)', '', sentence)
    

    # textss= re.sub(r'=+', '', texts)

    # # Remove elements like 3:4
    # textsss = re.sub(r'\d+:\d+', '', textss)
    # textsx = re.sub(r'#\d+', '', textsss)

    # # Remove elements like [3]
    # textax = re.sub(r'\[\d+\]', '', textsx)

    # textpg = re.sub(r'\[\d+\];', '', textax)

    # text = re.sub(r'\[Pg \d+\]', '', textpg)

    # text= re.sub(r'\(\d+\)|=+|\d+:\d+|#\d+|\[\d+\]|\[Pg \d+\];', '', sentence)



    # If there are multiple spaces or newline characters between the patterns you want to remove 
    # text1= re.sub(r'\s*\(?\d+\)?\s*|=+|\d+:\d+|#\d+|\[\d+\]|\[Pg \d+\];\s*', '', sentence)

    # # If your text spans multiple lines and you want to remove patterns across lines
    # text = re.sub(r'\(\d+\)|=+|\d+:\d+|#\d+|\[Pg\]|\[\d+\]|\[Pg \d+\];', '', text1, flags=re.MULTILINE)


# Remove patterns like "(123)", where 123 is a sequence of digits
    pattern1 = r'\(\d+\)'
    text1 = re.sub(pattern1, '', sentence)

    # Remove patterns like "=====" (multiple equals signs)
    pattern2 = r'=+'
    text2 = re.sub(pattern2, '', text1)

    # Remove patterns like "12:34" (time format)
    pattern3 = r'\d+:\d+'
    text3 = re.sub(pattern3, '', text2)

    # Remove patterns like "#123" (hash followed by digits)
    pattern4 = r'#\d+'
    text4 = re.sub(pattern4, '', text3)

    # Remove patterns like "[Pg]" (square brackets with "Pg" inside)
    pattern5 = r'\[Pg\]'
    text5 = re.sub(pattern5, '', text4)

    # Remove patterns like "[123]" (square brackets with digits inside)
    pattern6 = r'\[\d+\]'
    text6 = re.sub(pattern6, '', text5)


    # Remove patterns like "[Pg 123];" (square brackets with "Pg" and digits inside followed by a semicolon)
    pattern7 = r'\[Pg \d+\]'

    text7 = re.sub(pattern7, '', text6)

    pattern8 = re.sub(r'\s*\[\s*\d+\s*\]\s*', '', text7)

    

    text8 = re.sub(pattern7, '', pattern8, flags=re.MULTILINE)

    pattern9 = r'\[Pg [MDCLXVI]+\]'
    
    # Use the re.sub() function to remove all matches of the pattern
    text_final = re.sub(pattern9, '', text8)
    # Now, 'text_final' contains the cleaned text after all the specified operations



    words1=text_final.strip()

    words = words1.split()
    chunks = []
    current_chunk = []
    word_count = 0
    # print(words)
    for word in words:
        word_count += 1
        current_chunk.append(word)

        if word.endswith(('.',',' ,'?', '!','"')) or word_count >= max_words:
            chunks.append(' '.join(current_chunk))
            current_chunk = []
            word_count = 0

# 9755552076 Tanuja shetty || drink exchange

    if current_chunk:
        chunks.append(' '.join(current_chunk))
        
    return chunks

```
